# Titanic Tabulation

The Titanic is an amazing historical ship that was built starting in March, 1909. It took over 3 years to build, and was constructed at the Harland and Wolff shipyard in Belfast.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv("titanic_passengers.csv")
```

## 1. How many passengers were in first class?

```python
#Represents the number of passengers in each social class
pclass = titanic_df['Pclass'].value_counts()
plt.figure(figsize = (10,6), facecolor = 'aquamarine')
ax = plt.gca()
ax.set_facecolor('lightblue')
plt.title("Passenger Class")
pclass.plot.bar(edgecolor='black', rot = 360) 
plt.xlabel('Social Class')
plt.ylabel('Number of Passengers')
```

This graph shows the number of passengers with the respectful number towards what class they fall under. Compared to the other classes, the number of passengers that were in first class is less than third class, but more than second class. There are roughly a little over 200 people in first class.

## 2. What was the age distribution between passengers?

```python
#Displays the number of passengers to a certain age
ages = titanic_df['Age']
plt.figure(figsize = (10,6), facecolor = 'aquamarine')
ax = plt.gca()
ax.set_facecolor('lightblue')
plt.title("Passenger Age Distribution")
ages.plot.hist(bins=8, edgecolor='black')
plt.xlabel('Ages')
plt.ylabel('Number of Passengers')
```

This graph represents the ages of all of the passengers. There are 8 seprerate bars representing the ages by tens. The highest age in the graph being 80, and the lowest age being just slightly over 0, representing infants/babies.

## 3. What was the survival rate of females compared to men?

```python
#Shows percentile for survival and death rate based on gender
gender_casualties = titanic_df[['Survived','Sex']].value_counts()
plt.figure(figsize = (10,6), facecolor = 'lightblue')
ax = plt.gca()
plt.title("Percentage of Death/Survival")
gender_casualties.plot.pie(autopct='%1.1f%%', ax=ax, legend=True)
```

This graph represents the death and survival rates for men and women that were aboard the titanic. The number 1 indicating they survived, and the number 0 indicating they died. Compared to men, women on the titanic had a higer surival rate compared to them.

## 4. How many people survied/died from their respective class?

```python
#Displays a bar graph with two values with those who survived and those who didn't
survived = titanic_df.groupby("Pclass")["Survived"].sum()
died = titanic_df.groupby("Pclass")["Survived"].count() - survived
survival_class = pd.DataFrame({"Survived": survived, "Died": died})

colors = ['lightgreen', 'lightcoral']  
ax = survival_class.plot.bar(figsize=(10, 6), color=colors, edgecolor = 'black', rot = 360)

ax.set_title('Survival vs Death by Passenger Class')
ax.set_xlabel('Pclass')
ax.set_ylabel('Count')
ax.set_facecolor('lightblue')

fig = plt.gcf()  
fig.patch.set_facecolor('aquamarine')
```

This graph represent the amounf of people who died and survived based off of their respective class. Most people who were under class 3 had the highest number of deaths, as they were given less priority as the other two.

## 5.  Did class, age, and gender affect survival rates?

Comparing the data all together, we can assume that most of the casualties were of male passenger in 3rd class, primarily of ages 15 - 47.
